5 Write a program to reverse a string using a for loop.
-

```python
text = input("Enter a string: ")
revers = ""

for i in text:
    revers = i + revers

print("Reversed string:", revers)
```

Check whether a number is prime
-

```python
number= int(input("enter a number: "))

for i in range(2, number):
    if number % i == 0:
            print(number, "is not prime number")
            break
    else:                    
        print(number, "is a prime number")             
        break
```

10 Break and Continue
-
Print numbers from 1 to 20, but:
- Skip multiples of 3.
- Stop execution if number becomes greater than 15.

```python
for i in range(1, 21):

    if i > 15:
        break

    if i % 3 == 0:
        continue

    print(i)
```

Calculate factorial of a number using a while loop.
-

```python
number = int(input("enter a number: "))

fact = 1                      
i = 1
while i <= number:
    fact = fact * i   
    i = i + 1         

print("Factorial of", number, "is", fact)
```
