1 Student Grade Calculator
-
Write a function calculate_grade(marks) that takes marks as input and returns grade according to:

- 90 and above → A
- 80 -- 89 → B
- 70 -- 79 → C
- Below 70 → Fail

```python
def Student_Grade_Calculator(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "Fail"
```

```python
marks = int(input("enter marks: "))

print(Student_Grade_Calculator(marks)) 
```

3 Factorial Using Function
-
Write a function factorial(n) that returns the factorial of a number using a loop.

Example:
-
- factorial(5)
- Output → 120

```python
def factorial(number):
    fact = 1                      
    i = 1
    while i <= number:
        fact = fact * i   
        i = i + 1
    return fact
```

```python
number = int(input("enter a number: "))


print("Factorial of", number, "is", factorial(number))
```

6 Prime Number Checker
-
Write a function is_prime(n) that checks whether a number is a prime number.

```python

def  Prime_Number_Checker(number):
    if number <= 1:
        return False
    if number == 2:
        return True
    if number % 2 == 0:
        return False
    i = 3
    while i * i <= number:  
        if number % i == 0:  
            return False
        i = i + 2            
    return True
```

7 List Sum Function
-
Write a function list_sum(numbers) that takes a list as input and returns the sum of all numbers.

Example:
-
- [10, 20, 30]
- Output → 60

```python
def list_sum(number):
    total=0
    for num in number:
        total += num
    return total

print(list_sum([10, 20, 30])) 
```

8 Password Strength Checker
-
Write a function check_password(password) that checks:

- Minimum 8 characters required
- Must contain at least one digit
- Return:
&nbsp; "Strong" or "Weak"

```python
def check_password(password):
   
    if len(password) < 8:
        return "Weak"
    
    for character in password:          
        if character.isdigit():          
            return "Strong"
    else:
            return "Weak"
```

```python
print("Your password is ->", check_password("pass1234"))
```

9 Temperature Converter
-
Write a function celsius_to_fahrenheit(c) that converts temperature from Celsius to Fahrenheit.

Formula:
-
- F = (C × 9/5) + 32

```python
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

# Example
print(celsius_to_fahrenheit(25))  # Output → 77.0
```

10 Remove Duplicates from List
-
Write a function remove_duplicates(data) that removes duplicate elements from a list and returns a new list.

Example:
-
- [1,2,2,3,4,4]
- Output → [1,2,3,4]

```python
def remove_duplicates(data):
    return list(set(data))
    
print(remove_duplicates([1, 2, 2, 3, 4, 4]))
```
