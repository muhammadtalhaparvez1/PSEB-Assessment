Q2. Swap Two Variables
-
Write a Python program that takes two numbers in two variables and swaps them.

Example Input:
-
Enter first number: 5
Enter second number: 10

Expected Output:
-
After swapping:
First number: 10
Second number: 5



```python

num1: int =5
num2: int =10
print('Before Swapping')
print('first number : ',num1)
print('second number :',num2)

temp = num1
num1 = num2
num2 = temp
print('After Swapping')
print('first number : ',num1)
print('second number :',num2)
```

Q3. Swap Two Variables
-
Write a Python program that takes two numbers in two variables and swaps them without using third variable.

Example Input:
-
Enter first number: 5
Enter second number: 10

Expected Output:
-
After swapping:
First number: 10
Second number: 5

```python

print("\nTask 3")
num1: int = 5
num2: int = 10
print('Before Swapping')
print('First number : ',num1)
print('Second number : ',num2)
num1  = num1 + num2
num2  = num1 - num2
num1 = num1 - num2
print('After Swapping')
print('First number : ',num1)
print('Second number : ',num2)
```
