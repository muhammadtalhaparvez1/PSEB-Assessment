1 Write a Python program that can do the following:
   -

You have $50.

You buy an item that costs $15 with a 3% tax.

Print how much money you have left, after purchasing the item.

```python


money= 50
print("Total Amount:", money)

item_price =15

tax=0.03 #3%

total_cost_plus_tax= item_price +( item_price * tax)

print("Total cost including tax : ", total_cost_plus_tax)

money_left = money - total_cost_plus_tax

print("Amount left: " , money_left)
```

2 Ask the user how many days until their birthday
   -
Print an approx. number of weeks until their birthday<br>
1 week is = to 7 days.<br>

```python
days_left_in_bd= int(input("Enter days left in your birth day: " ))

days_left= days_left_in_bd/7
days_left= round(days_left,3)
print("total weeks left in your birth day are: ", days_left)
```

3 Create a variable “Grade” holding an integer between 0 to100. Code to print
the letter grade of the number grade variable.
-
Grades:
- A = 90 - 100
- B = 80 - 89
- C = 70-79
- D = 60 - 69
- F = 0 – 59

```python
Grade = int(input("Input your grades: "))

if Grade >= 90 and Grade <= 100:
    print("A")
elif Grade >= 80 and Grade <= 89:
    print("B")
elif Grade >= 70 and Grade <= 79:
    print("C")
elif Grade >= 60 and Grade <= 69:
    print("D")
elif Grade >= 0 and Grade <= 59:
    print("F")
```
