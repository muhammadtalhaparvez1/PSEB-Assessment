## 3 List methods

1. Append a new fruit to the list using `append()`. Print the updated list.  
2. Remove one fruit from the list using `remove()` (or `pop()`). Print the updated list again.  
3. Print the length of the list using `len()`.  

## Example Output (Walkthrough)

```python
fruits_input = input("Enter three of your favorite fruits: ")
```

```python
fruit_list = fruits_input.split(" ")

print("Your list of fruits:", fruit_list)
```

```python
print("First fruit:", fruit_list[0])
print("Sexond fruit:", fruit_list[1])
print("Last fruit:", fruit_list[-1])
print("First Two fruit:", fruit_list[0:2])

```

```python
fruit_list.append('strawberry')
print("Updated list after adding a new fruit:", fruit_list)
```

```python
fruit_list.remove('banana') 
print("Updated list after removing a fruit:", fruit_list)
```

```python
print("Length of the list is now:", len(fruit_list))
```

## 3 Simple Arithmetic on a List of Numbers

- Prompt the user to enter 5 numbers (e.g., `"10 20 30 40 50"`) in one line, separated by spaces.  
- Convert each input into an integer and store all in a list called `numbers_list`.  
- Print the sum of the numbers, and the average (`sum / length of the list`).

## Example Output (Walkthrough)

Last 3 letters: ['D', 'o', 'e']<br>

```python
full_name = input("Enter your full name: ")
```

```python
print("Full name: ",full_name)
letters_list = list(full_name.replace(" ", ""))
print("Letter list: ", letters_list)
print("First 3 letters:", letters_list[:3])

print("Last 3 letters:", letters_list[-3:])
```

```python
#first three in upper case 

for i, letter in enumerate(letters_list):
    if i < 3:
        letters_list[i] = letter.upper()
    else:
        letters_list[i] = letters_list[i].lower() 
print(letters_list)
```

```python
#joining name back together

modified_name = "".join(letters_list)
print("Modified name:", modified_name)

```

```python
numbers_list = []

for i in range(5):
    a = int(input(f"enter a number{i}"))
    numbers_list.append(a)
```

```python
numbers_list
```

```python
total = sum(numbers_list)
average = total / len(numbers_list)

print("Sum of your numbers:", total)
print("Average of your numbers:", average)
```
